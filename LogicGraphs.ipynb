```python
class LogicModel:
    def __init__(self, inputs, activities, outputs, outcomes):
        self.inputs = inputs
        self.activities = activities
        self.outputs = outputs
        self.outcomes = outcomes

    def display(self):
        print("Logic Model:")
        print(f"Inputs: {self.inputs}")
        print(f"Activities: {self.activities}")
        print(f"Outputs: {self.outputs}")
        print(f"Outcomes: {self.outcomes}")

# Create a logic model
model = LogicModel(
    inputs=["Funding", "Staff"],
    activities=["Train staff", "Run workshops"],
    outputs=["50 trained staff"],
    outcomes={"Short-term": "Increased awareness"}
)

model.display()
```

```python
import simpy

def workshop_process(env, attendees, outcomes):
    print(f"Workshop starts at {env.now}")
    yield env.timeout(5)  # Simulate 2 days for the workshop
    outcomes.append(f"{attendees} participants completed the workshop at day {env.now}")
    print(f"Workshop ends at {env.now}")

# Initialize the environment
env = simpy.Environment()

# Define parameters
attendees = 50
outcomes = []

# Run the simulation
env.process(workshop_process(env, attendees, outcomes))
env.run()

# Display results
print("Outcomes from simulation:")
print(outcomes)
```
